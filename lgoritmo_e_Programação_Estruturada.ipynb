```python
%%writefile aula_ads.c
#include <stdio.h>

int main() {
    float salario;
    int possuiDividas;

    printf("--- SISTEMA DE CREDITO --- \n");
    printf("Digite o seu salario: ");
    scanf("%f", &salario);

    printf("Possui dividas? (1-Sim / 0-Nao): ");
    scanf("%d", &possuiDividas);

    if (salario >= 3000) {
        if (possuiDividas == 0) {
            printf("\nStatus: APROVADO\n");
        } else {
            printf("\nStatus: EM ANALISE\n");
        }
    }
    else {
        if (possuiDividas == 0) {
            printf("\nStatus: EM ANALISE\n");
        } else {
            printf("\nStatus: NEGADO\n");
        }
    }
    return 0;
}
```
